```python
import pandas as pd
import sqlalchemy as sa


# Define the query
query = sa.text("""
SELECT DISTINCT 
    n.network_id,
    s.native_id, 
    m.station_name, 
    s.min_obs_time, 
    s.max_obs_time, 
    m.lat, 
    m.lon, 
    m.elev, 
    m.station_id,
    m.mod_time
FROM meta_history AS m
JOIN meta_station AS s ON m.station_id = s.station_id
JOIN meta_network AS n on n.network_id = s.network_id
WHERE n.network_id = :network_id
""")

# Execute query and read into DataFrame
with engine.begin() as conn:
    df = pd.read_sql(query, conn, params={"network_id": 10})

# Save to Excel
df.to_excel("29Jan_Network_10_AGRI_metadata.xlsx", index=False)

```
